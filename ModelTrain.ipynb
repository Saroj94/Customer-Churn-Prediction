```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.ensemble import VotingClassifier,StackingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix, recall_score,precision_score,f1_score,roc_auc_score
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
##voting classifier
voting_clf=VotingClassifier(estimators=[
    ('GB',GradientBoostingClassifier()),
    ('AB',AdaBoostClassifier()),
    ('LR',LogisticRegression()),
    ('SVM',SVC())
    ])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
# Custom label encoder wrapper
class PretrainedLabelEncoderWrapper(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.encoders = {}

    def fit(self, X, y=None):
        for col in X.columns:
            le = LabelEncoder()
            le.fit(X[col])
            self.encoders[col] = le
        return self

    def transform(self, X):
        X_encoded = X.copy()
        for col in X.columns:
            le = self.encoders[col]
            X_encoded[col] = le.transform(X[col])
        return X_encoded

    def inverse_transform(self, X):
        X_decoded = X.copy()
        for col in X.columns:
            le = self.encoders[col]
            X_decoded[col] = le.inverse_transform(X[col])
        return X_decoded
```

```python
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer
import pickle
# Define columns
transformed_cols = ['MonthlyCharges', 'TotalCharges', 'tenure']
categorical_features = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
                        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','Contract', 'PaymentMethod']
label_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']


# Define transformers
numeric_transformer = Pipeline([
    ('power', PowerTransformer(method='yeo-johnson'))
])

categorical_transformer = Pipeline([
     ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
 ])

label_encoder_transformer = Pipeline([
    ('label', PretrainedLabelEncoderWrapper())
])

# Build preprocessor
preprocessor = ColumnTransformer([
    ('num', numeric_transformer, transformed_cols),
      ('cat', categorical_transformer, categorical_features),
     ('labels_cols', label_encoder_transformer, label_cols)
],remainder='passthrough',sparse_threshold=0.0)

#Transform training independent features
X_train_processed = preprocessor.fit_transform(X_train)

# Apply SMOTE to the transformed training set(X_train and y_train) to balance the imbalance class
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X_train_processed, y_train)

# Train the voting classifier on the resampled data
voting_clf.fit(X_resampled, y_resampled)

inference_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', voting_clf)
])


inference_pipeline.predict()

#saving the preprocessing model and machine learning model for future prediction
with open("VotingClassifiers.pkl","wb") as filesaved:
    pickle.dump(inference_pipeline,filesaved)
```
